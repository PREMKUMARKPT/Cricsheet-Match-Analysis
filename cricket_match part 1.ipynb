```python
import json
import pandas as pd
import os


# Folder where all JSON files are stored i stored
folder_path = "ipl"  # Update if necessary

# List to store all match data
all_matches = []

# Loop through all JSON files in the folder
for file_name in os.listdir(folder_path):
    if file_name.endswith(".json"):  # Process only JSON files
        file_path = os.path.join(folder_path, file_name)

        # Load JSON file
        with open(file_path, "r") as file:
            data = json.load(file)


        # Extract match data
        for inning in data["innings"]:
            team = inning["team"]
            for over_details in inning["overs"]:
                over = over_details["over"]
                for delivery in over_details["deliveries"]:
                    all_matches.append({
                        "match_id": file_name.split(".")[0],  # Extract match ID from filename
                        "match_type": data["info"]["match_type"],
                        "team": team,
                        "over": over,
                        "batter": delivery["batter"],
                        "bowler": delivery["bowler"],
                        "non_striker": delivery["non_striker"],
                        "runs_batter": delivery["runs"]["batter"],
                        "runs_extras": delivery["runs"]["extras"],
                        "runs_total": delivery["runs"]["total"],
                    })


# Convert all match data into a single DataFrame
dfipl = pd.DataFrame(all_matches)


# Save to CSV
csv_path = "ipl_match.csv"
dfipl.to_csv(csv_path, index=False)


print(f"✅ CSV file saved at: {csv_path}")

```
