```python
# Team with Highest Win Percentage (Test)

mycursor = mydb.cursor()

# Corrected syntax with parentheses and using parameterized query
mycursor.execute("""
SELECT team,
       COUNT(DISTINCT match_id) AS wins,
       ROUND(COUNT(DISTINCT match_id) * 100.0 /
            (SELECT COUNT(DISTINCT match_id) FROM test_match), 2) AS win_percentage
FROM test_match
GROUP BY team
ORDER BY win_percentage DESC
LIMIT 1;


""")

myresult = mycursor.fetchall()

for x in myresult:
    print(x)
```
